### Some commonly used Matplotlib.pyplot **Functions**

- **bar()**: Make a bar plot
- **scatter()**: A scatter plot of y vs x
- **boxplot()**: Make a box and whisker plot
- **hist()**: Plot a histogram
- **semilogx()**: Make a plot with log scaling on the x axis
- **semilogy()**: Make a plot with log scaling on the y axis

- **xcorr()**: 	Plot the cross correlation between x and y
---

- **figure()**: Create a new figure, or activate an existing figure
- **legend()**:Place a legend on the Axes
- **plot()**: Plot y versus x as lines and/or markers
- **savefig()**: Save the current figure

- **xlabel()**: Set the label for the x-axis

- **ylabel()**: Set the label for the y-axis

```python
#lets plot now box-whisker's plot (boxplot)
#boxplot helps to identify the outliers
plt.boxplot(data['Brain Weight(grams)'])
plt.show()
```

```python
#finding outlier points
#Upper limit = Q3+1.5(IQR)
#Lower limit = Q1-1.5(IQR)

q1 = data['Brain Weight(grams)'].quantile(0.25)  
q3 = data['Brain Weight(grams)'].quantile(0.75)

IQR = q3-q1

upper = q3+1.5*IQR
lower = q1-1.5*IQR

upper_out = data.loc[(data['Brain Weight(grams)']>upper)]
lower_out = data.loc[(data['Brain Weight(grams)']<lower)]

outliers = pd.concat([lower_out, upper_out]) #concatinating DataFrames
outliers
```

```python
#Droping the outlier rows from actual datasets
data_modified = data.drop(index=outliers.index, axis=0)
data_modified
```

```python
#Lets call a datset from seaborn library to understand some features
#Lets import 'tips' dataset. This dataset is having information of tip given based on total bill paid, sex, day of the week, time of the day, size of groups(people)
#and whether the person is smoker or not.

tips = sns.load_dataset('tips') # This will call the tips DataFrame into tips object
tips.head()
```

```python
tips.shape
```

```python
tips.info()
```

```python
#Now lets plot some relationships and try to understand the datasets
#Initialize the matplotlib figure and FacetGrid object.

sns.FacetGrid(tips) #It will just create null grid to plot the graphs
plt.show()
```

```python
# To draw a plot on every facetgrid, pass a function and the name of one or more columns in the dataframe and map it with kind of plot you want
# Lets create an FacetGrid object (e.g. g) and pass arguments
# Before that lets plot a simple scatterplot

sns.scatterplot(x=tips['total_bill'], y=tips['tip'])
plt.show()
```

```python
#Now do the same thing with facetgrid

g = sns.FacetGrid(tips)
g.map(sns.scatterplot, 'total_bill', 'tip')
plt.show()
```

```python
#Now we can subset the scatterplot based on other conditions from the same DataFrame, e.g. time and sex

g = sns.FacetGrid(tips, col='time', row='sex')
g.map_dataframe(sns.scatterplot, 'total_bill', 'tip') # This will give all combinations of 'total_bill' vs 'tip' based on 'time' and 'sex'.
plt.show()
```

```python
#Lets plot histogram of total bill

g = sns.FacetGrid(tips, col="time",  row="sex")
g.map_dataframe(sns.histplot, x="total_bill", binwidth=2) #binwidth to set the range of bars
plt.show()
```

```python
#Set axis level for variables
g = sns.FacetGrid(tips, col="time", row="sex")
g.map_dataframe(sns.histplot, x="total_bill", binwidth=2)
g.set_axis_labels("Total bill", "Count")
plt.show()
```

```python
g = sns.FacetGrid(tips, col="time", hue="sex") #hue differentiate the categories of a categorical variables
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip") #we can pass hue either while defining the grid or can pass while plotting the graph
g.set_axis_labels("Total bill", "Tip")
g.add_legend()
plt.show()
```

```python
g = sns.FacetGrid(tips, row='sex', col='time')
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
g.set_axis_labels("Total bill", "Tip")
g.add_legend()
plt.show()
```

```python
#Plot lmplot
sns.lmplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
#subsetting lmplot
g = sns.lmplot(data=tips, x='total_bill', y='tip', col='time', row='sex', height=3) #height is to set the height of the graphs
g.set_axis_labels("Total bill", "Tip")
g.add_legend()
plt.show()
```

```python
#Export graph
g = sns.lmplot(data=tips, x='total_bill', y='tip', col='time', row='sex', height=3) #height is to set the height of the graphs
g.set_axis_labels("Total Bill", "Tip")
g.add_legend()
plt.plot()
plt.savefig('tips.png', dpi=150)
```

```python
# pairplot is will compare each numerical variable with all other numerical variables.
# This helps to quickly visualize the data and understand the relaionship with variables.

sns.pairplot(tips)
plt.show()
```
